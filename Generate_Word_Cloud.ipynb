# Generate Word Cloud

This notebook uses the Python Word Cloud module to visualise the contents of a collection of text. It is based on the article [Simple word cloud in Python](https://towardsdatascience.com/simple-wordcloud-in-python-2ae54a9f58e5).

GitHub Repository: https://github.com/amueller/word_cloud

Matplotlib Colormaps: https://matplotlib.org/3.2.1/tutorials/colors/colormaps.html

#### Import packages

```python
import pandas as pd 

from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt
%matplotlib inline
```

#### Read CSV as Dataframe

```python
data = pd.read_csv('../software_data.csv')
```

#### Display Dataframe

```python
data.head()
```

#### Convert Data to Text

```python
text = (' '.join(data['abstract']))
```

#### Clean Text

```python
text = text.replace('\n', '')
text
```

#### Define a function to plot word cloud

```python
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off");
```

```python
wordcloud = WordCloud(width = 3000, 
                      height = 2000, 
                      random_state=1, 
                      background_color='black', 
                      colormap='Set2', 
                      collocations=False, 
                      stopwords = STOPWORDS
                     ).generate(text)
```

#### Plot Word Cloud

```python
plot_cloud(wordcloud)
```

#### Save the Word Cloud

```python
wordcloud.to_file('../wordcloud.png')
```
